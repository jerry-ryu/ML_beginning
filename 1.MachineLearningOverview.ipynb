# 1. 머신러닝 개요
--------
## 1-1. 머신러닝이란?
> 머신러닝은 데이터에서부터 학습하도록 컴퓨터를 프로그래밍하는 과학이다
(애플리케이션을 수정하지 않고 패턴을 학습 -> 결과 예측, 유동적이고 대량의 데이터 환경에 유리하다.)

> 어떤 작업 T에 대한 컴퓨터 프로그램의 성능을 P로 측정했을 때 경험 E로 인해 성능이 향상됐다면, 이 컴퓨터 프로그램은 작업 T와 성능 측정 P에 대해 경험 E로 학습한 것이다.

## 1-2. 머신러닝의 분류-1
1. 지도학습(Supervised Learning)   
지도학습은 레이블된 훈련 데이터에서 모델을 학습하여 본 적 없는 미래 데이터에 대해 예측을 만드는 것.   
 - 지도(supervised): 희망하는 출력신호(label)이 있는 일련의 샘플
    
  ex. 분류(과거의 관측을 기반으로 새로운 샘플의 범주형 클래스 레이블을 예측), 회귀(예측변수와 연속적인 반응변수가 주어졌을 때, 출력 값을 예측하는 두 변수 사이의 관계를 찾음), 추천시스템, 시각/음성감지/인지, 텍스트 분석, NLP   
2.비지도 학습(Unsupervised Learning)   
레이블되지 않거나 구조를 알 수 없는 데이터가 어떻게 이루어져있는지 스스로 알아내는 방법   
ex. 클러스터링(군집화), 시각화, 차원축소, 연관 규칙 학습, 이상치/특이치 탐지   
3.준지도학습(Semi - Supervised Learning)   
라벨링된 데이터와 라벨링되지 않은 데이터를 모두 훈련에 사용, 소량의 labeled data에는 지도 학습을 대용량 unlabeled data에는 비지도 학습을 사용하여 단지 label을 맞추는 모델에서 벗어나 데이터 자체의 본질적인 특성을 모델링하는 것이 목표  
 - 'data labeling'에 많은 자원과 비용 소모 -> 대안으로 준지도학습 등장
4. 강화학습(Reinforcement Learning)   
어떤 환경 안에서 정의된 에이전트가 현재의 상태를 인식하여, 선택 가능한 행동들 중 보상을 최대화하는 행동 혹은 행동 순서를 선택하는 방법

## 1-3. 머신러닝의 분류-2
1. 배치학습(Batch Learning)  
한번에 모든 훈련 데이터를 학습시키는 방법으로, 시간과 자원을 많이 소모, 주로 오프라인에서 수행된다.(== Offline Learning)   
2.온라인학습(Online Learning)   
학습이 끝나 제품화가 된 모델에 대하여, 미니 배치 단위의 데이터를 주입하여 모델을 훈련시키는 방법.(== 점진적 학습, Increment Learning)  
데이터 양이 지나치게 커서, 컴퓨터의 메모리로 감당되지 않은 경우에도 온라인 학습이 사용된다.(== 외부 메모리 학습, Out-of-core Learning)

## 1-4. 머신러닝의 분류-3
1. 사례기반 학습(Instance based Learning)  
시스템이 단순히 여러 사례(instance)를 메모리에 저장함으로써 학습, 훈련 인스턴스들은 입력된 인스턴스와 가장 비슷한 인스턴스를 찾기 위해 사용(유사도 측정)   
ex. KNN 분류   
2.모델 기반 학습(Model-based Learning)   
여러 샘플 데이터들을 통해 일반화 시키기 위해서 모델을 만들어 예측에 사용
  1. 데이터의 형태 가정
  2. 모델의 학습 목표 수식화
  3. 실제 데이터로 모델 학습(최적화)
  4. 모델 평가

##1-5. 머신러닝이 당면한 문제
 - 충분하지 않은 훈련 데이터
 - 대표성 없는 훈련 데이터:   
 (샘플링 잡음(우연에 의한 대표성 없는 데이터, sampling noise), 샘플링 편향(표본 추출 방법으로 인한 문제, sampling bias)으로 인한 문제)

-  낮은 품질의 데이터:   
에러, 이상치, 잡음으로 데이터의 품질이 저하될 수 있으므로, 데이터의 정제가 필요하다   

- 관련없는 특성:   
특성 선택: 가지고 있는 특성 중에서 훈련에 가장 유용한 특성 선택   
특성 추출: 특성을 결합하여 더 유용한 특성 생성   

- 훈련 데이터 과소/과대적합:   
모델이 너무 복잡해져서 학습할 필요없는 데이터의 구조 학습 -> 과대적합(Overfitting)    모델이 너무 단순해서 데이터의 구조를 학습 X -> 과소적합(Underfitting)

\right] $$

\left[
\begin{matrix}
    y^{(1)} \\
    \vdots\\
    y^{(150)}
\end{matrix}
\right] $$

numpy 인덱싱(Indexing)
1. 단일 값 추출: 원하는 위치의 인덱스 값을 지정하면 해당 위치의 데이터가 반환된다 

````python
# 1차원 ndarray 단일 값 추출
array1 = np.arange(1,10)
print(array1)
print(array1[7])
print(type(array1[7]))
print()

# 2차원 ndarray 단일 값 추출
array2 = np.arange(0,9).reshape(3,3)
print(array2)
print(array2[2,0])
print(type(array2[2,0]))
````

   
   2. 슬라이싱(slicing): 연속된 인덱스 상의 ndarray를 추출하는 방식.   
   *시작 인덱스 : 종료 인덱스 -> 시작 인덱스 ~ 종료 인덱스 -1 ndarray 반환   
   *생략된 값은 맨 처음/끝 인덱스로 간주한다

````python
# 1차원 ndarray 슬라이싱
array1 = np.arange(1,10)
print(array1)
print(array1[3:8])
print(type(array1[3:8]))
print()

# 2차원 ndarray 슬라이싱
array2 = np.arange(0,9).reshape(3,3)
print(array2)
print(array2[:,:2])
print(type(array2[:,:2]))
````

   
   3. 팬시 인덱싱(Fancy Indexing): 일정한 인덱싱 집합을 리스트 또는 ndarray 형태로 지정해 해당 위치에 있는 데이터의 ndarray 반환

````python
# 1차원 ndarray 팬시 인덱싱
array1 = np.arange(1,10)
print(array1)
print(array1[[1,5]])
print(type(array1[[1,5]]))
print()

# 2차원 ndarray 팬시 인덱싱
array2 = np.arange(0,9).reshape(3,3)
print(array2)
print(array2[[1,2],2])
print(type(array2[[1,2],2]))
````

   
   4. 불린 인덱싱(Boolean Indexing): 특정 조건에 해당하는지 여부인 True/False 값 인덱싱 집합을 기반으로 True에 해당하는 인덱스 위치에 있는 데이터의 ndarray 반환

````python
# 1차원 ndarray 불린 인덱싱
array1 = np.arange(1,10)
print(array1)
print(array1[array1>=5])
print(type(array1[array1>=5]))
print()

# 2차원 ndarray 팬시 인덱싱
array2 = np.arange(0,9).reshape(3,3)
print(array2)
print(array2[array2 != 2])
print(type(array2[[1,2],2]))
````

numpy 정렬:   
 - np.sort(): 원 행렬은 그대로 유지, 정렬된 행렬 반환
 - ndarray.sort(): 원 행렬을 정렬한 형태로 변환, None 반환
 - np.argsort(): 정렬 행렬의 원본 행렬 인덱스를 ndarray형으로 반환

````python
org_array = np.array([3,1,9,5]) # 원 행렬
print('원본 행렬:', org_array)
print()

#np.sort() 정렬
sort_array1 = np.sort(org_array)
print('반환된 정렬 향렬:', sort_array1)
print('원본 행렬:', org_array)
print()

#ndarray.sort() 정렬
org_array = np.array([3,1,9,5]) # 원 행렬
sort_array2 = org_array.sort()
print('반환된 정렬 향렬:', sort_array2)
print('원본 행렬:', org_array)
print()

#np.argsort
org_array = np.array([3,1,9,5]) # 원 행렬
sort_indices = np.argsort(org_array)
print(type(sort_indices))
print('행렬 정렬 시 원본 행렬의 인덱스:', sort_indices)
````

````python
# argsort의 활용
# arr1을 기준으로 arr2 정렬하기
arr1 = np.array([3,1,9,5]) 
arr2 = np.array([1,2,3,4])

s = np.argsort(arr1) # arr1을 정렬하면서 원본 인덱스 반환
print(s)
print(arr2[s]) # 원본 행렬 인덱스로 arr2 인덱싱하여, arr1을 기준으로 arr2 정렬
````

 - 내림차순 정렬하기
  - np.sort(x)[::-1](mirror view)
  - x[np.argsort(-x)]

  *arr[A:B:C] == A부터 B까지 C의 간격으로 배열을 만든다
  *arr[::-1] == 처음부터 끝까지 -1의 간격(역순)으로 배열을 만든다   

- axis(축)의 개념
  - 1차원 배열: axis0 존재   
    ex. [1,2,3] 에서 axis0을 따라 진행: 1 -> 2 -> 3
  - 2차원 배열: axis0, axis1 존재   
    ex.[[1,2],[3,4]] 에서 axis0을 따라 진행 [1,2]->[3,4]   
    ex.[[1,2],[3,4]] 에서 axis1을 따라 진행 1 -> 2 , 3 -> 4
  -3차원 배열: axis0, axis1, axis3 존재
    ex. [[[1,2],[3,4]],[[5,6],[7,8]]] 에서 axis0을 따라 진행 [[1,2],[3,4]] ->[[5,6],[7,8]]   
    ex. [[[1,2],[3,4]],[[5,6],[7,8]]] 에서 axis1을 따라 진행 [1,2] -> [3,4], [5,6] -> [7,8]   
    ex. [[[1,2],[3,4]],[[5,6],[7,8]]] 에서 axis3을 따라 진행 1->2,3->4,5->6,7->8

- 다차원 정렬   
  - 2차원 배열 정렬:   
      - axis 0으로 정렬 세로 기준으로 정렬
      - axis 1으로 정렬 가로 기준으로 정렬

````python
# 내림차순 정렬
org_array = np.array([3,1,9,5])
print('내림차순 정렬:')
mirror_view_array = np.sort(org_array)[::-1]
print(mirror_view_array)
argsort_array = np.sort(org_array)[np.argsort(-np.sort(org_array))]
print(argsort_array)
print()
print()


# axis
arr_1D = np.array([1,2,3,4])
print('1차원:\n', arr_1D)
print('axis0 sum\n:', np.sum(arr_1D, axis = 0))
print()

arr_2D = np.array([[1,2],[3,4]])
print('2차원:\n', arr_2D)
print('axis0 sum:\n', np.sum(arr_2D, axis = 0))
print('axis1 sum:\n', np.sum(arr_2D, axis = 1))
print()

arr_3D = np.array([[[1,2],[3,4]],[[5,6],[7,8]]])
print('3차원:\n', arr_3D)
print('axis0 sum:\n', np.sum(arr_3D, axis = 0))
print('axis1 sum:\n', np.sum(arr_3D, axis = 1))
print('axis2 sum:\n', np.sum(arr_3D, axis = 2))
print()
print()


# 다차원 정렬
array2d = np.array([[8,12],
                    [7,1]])
print('원 행렬\n', array2d)
sort_array2d_axis0 = np.sort(array2d, axis = 0) # 세로로 정렬
print('axis0 sort:\n', sort_array2d_axis0)
sort_array2d_axis1 = np.sort(array2d, axis = 1) # 가로로 정렬
print('axis1 sort:\n', sort_array2d_axis1)
````

````python
def get_category(age):
  '''나이에 따른 분류 함수'''
  cat = ''
  if age <= 15 : cat = 'Child'
  elif age <= 30: cat = 'Young Adult'
  else: cat = 'Adult'

  return cat;

# apply(), DF에 함수 적용하기
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : get_category(x))
titanic_df[['Age','Age_cat']].head(10)
````

9. GroupBy

````python
#groupby 예제
df = pd.DataFrame({
    'city': ['부산', '부산', '부산', '부산', '서울', '서울', '서울'],
    'fruits': ['apple', 'orange', 'banana', 'banana', 'apple', 'apple', 'banana'],
    'price': [100, 200, 250, 300, 150, 200, 400],
    'quantity': [1, 2, 3, 4, 5, 6, 7]
})
df
````

````python
# 같은 값을 하나로 묶기, groupby()
df.groupby('city').mean() # city column의 값으로 DF 재조직 후 평균값 구하기
````

````python
# 같은 값을 하나로 묶기, groupby()
df.groupby(['city','fruits']).mean() # city, fruits의 값으로 column의 값으로 DF 재조직 후 평균값 구하기
````

````python
# 같은 값을 하나로 묶기, groupby()
df.groupby(['city','fruits']).mean().unstack() # city, fruits의 값으로 column의 값으로 DF 재조직 후 평균값 구하기 (보기좋게 unstack())
````

````python
# 묶인 값 확인하기 get_group()
df.groupby('city').get_group('부산') # city column의 값으로 DF 재조직 후 그룹명이 '부산'인 데이터 출력
````

````python
# 그룹 순회
grouped_df = df.groupby(['city','fruits'])
for key, group in grouped_df:
  print(key)
  print(len(group))
````

````python
# 그룹 사이즈, size()
df.groupby('city').size()
````

````python
# 그룹에 aggregation 적용
df.groupby('city').agg({'price':'mean','quantity':'median'}) # city column의 값으로 DF 재조직 후 agg로 통계값 얻기
````
